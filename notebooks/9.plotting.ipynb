# Data visualization

```python
import numpy as np
import pandas as pd
```

```python
data_raw = pd.read_csv('../data/curated/combined_data_div.csv')
data_average = pd.read_csv('../data/curated/combined_data_average.csv')
label = pd.read_csv('../data/curated/combined_label.csv')
data_original = pd.read_csv('../data/curated/combined_data.csv')
```

```python
# classify data_average by label
data_average_FYZ = data_average[label['FYZ-IgM'] == 1]
data_average_jialiu = data_average[label['甲流'] == 1]
data_raw_FYZ = data_raw[label['FYZ-IgM'] == 1]
data_raw_jialiu = data_raw[label['甲流'] == 1]
data_original_FYZ = data_original[data_original['FYZ-IgM'] == 1.0]
data_original_jialiu = data_original[data_original['甲流'] == 2.0]
```

```python
data_average_jialiu['cumulative_mean_EO%']
```

```python
data_average_FYZ['cumulative_mean_EO%']
```

```python
# draw the boxplot of cumulative_mean_EO% for jialiu and fyz in one figure
import matplotlib.pyplot as plt
import seaborn as sns

# 设置输出语言为中文
fig, ax = plt.subplots()
sns.boxplot(data=[data_average_FYZ['cumulative_mean_MPV'][1:], data_average_jialiu['cumulative_mean_MPV'][1:]], ax=ax)
ax.set_xticklabels(['FYZ-IgM', 'Influenza A'])
ax.set_ylabel('cumulative mean of EO%')
fig.savefig('../data/plots/boxplot_1st_feature_importance.png')


```

```python
fig, ax = plt.subplots()
sns.boxplot(data=[data_original_FYZ['MPV'][1:], data_original_jialiu['MPV'][1:]], ax=ax)
ax.set_xticklabels(['FYZ-IgM', 'Influenza A'])
ax.set_ylabel('EO%')
```
